## Research Background

The increasing demand for data-driven applications has created a significant need for high-quality synthetic tabular data that preserves the statistical characteristics of real-world datasets while protecting sensitive information. Conventional anonymization techniques often fail to provide adequate privacy protection and may significantly reduce data utility.

This research proposes a unified framework that integrates statistical data generation techniques with deep generative models to produce high-fidelity synthetic tabular datasets while incorporating privacy-preserving mechanisms. The framework aims to achieve an optimal balance between data utility, statistical fidelity, and privacy protection.

## Research Objectives

The primary objectives of this research are:

1. Develop a unified framework integrating statistical and deep learning approaches for synthetic tabular data generation.

2. Compare traditional statistical generators with modern deep generative models.

3. Incorporate privacy-preserving mechanisms into the synthetic data generation process.

4. Evaluate synthetic datasets using fidelity, utility, and privacy metrics.

5. Identify the most effective model for generating high-quality privacy-preserving synthetic tabular data.

## Evaluation Metrics

The generated synthetic datasets will be evaluated using three categories of metrics.

### Fidelity

- Kolmogorov–Smirnov (KS) Test
- Wasserstein Distance
- Jensen–Shannon Divergence
- Correlation Preservation

### Utility

- Accuracy
- Precision
- Recall
- F1-Score
- ROC-AUC

### Privacy

- Membership Inference Attack Success Rate
- Attribute Disclosure Risk
- Re-identification Risk
- Differential Privacy Budget (ε)
